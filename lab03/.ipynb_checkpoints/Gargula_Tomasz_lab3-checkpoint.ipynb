Początkowo chciałem rozwiązać ten problem funkcyjnie – jednak z jakiegoś powodu kod funkcyjny wykonuje się 8-10 razy dłużej od analogicznego kodu imperatywnego. Zdecydowałem się jednak zostawić zapis funkcyjny, bo moim zdaniem wygląda ładnie.

```julia
# L(X, k, x) = reduce(*, [(x - xi) / (X[k] - xi) for (i, xi) in enumerate(X) if i != k])
# P(X, Y, x) = reduce(+, [y * L(X, k, x) for (k, y) in enumerate(Y)]);
```

```julia
function L(X, k, x)
    result = 1
    for (i, xi) in enumerate(X)
        if i != k
            result *= (x - xi) / (X[k] - xi)
        end
    end
    return result
end

function P(X, Y, x)
    result = 0
    for (k, y) in enumerate(Y)
        result += y * L(X, k, x)
    end
    return result
end;
```

```julia
Lagrange(x) = P(X, Y, x);
```

## Zadanie 2

```julia
function coefficients(X, Y)
    n = size(X, 1)
    b = Array{Float64}(undef, n, n)
    for i in 1:n
        b[i, 1] = Y[i]
    end
    for i in 1:n
        for j in 2:i
            b[i, j] = (b[i,j-1] - b[i-1, j-1]) / (X[i] - X[i - j + 1])
        end
    end
    return [b[i,i] for i in 1:n]
end;
```

```julia
function factor(X, n, x)
    p = Array{Float64}(undef, n)
    p[1] = x - X[1]
    for k in 2:n
        p[k] = p[k-1] * (x - X[k])
    end
    return p
end;
```

```julia
function PN(X, Y, x, b)
    n = size(b, 1)
    p(x) = factor(X, n, x)
    result = 0
    for i in 2:n
        result += b[i] * p(x)[i-1]
    end
    return Y[1] + result
end;
```

```julia
NewtonPrep = coefficients(X, Y)
Newton(x) = PN(X, Y, x, NewtonPrep);
```

## Zadanie 4

```julia
using DataFrames
using Statistics;
```

```julia
lagrange = Array{Float64}(undef, 0)
newtonPrep = Array{Float64}(undef, 0)
newton = Array{Float64}(undef, 0)
polynomialsPrep = Array{Float64}(undef, 0)
polynomials = Array{Float64}(undef, 0)

for s in 1000:1000:10000
    X = 1:s
    Y = rand(s)
    for i in 1:10
        append!(newtonPrep, @elapsed (NewtonPrep = coefficients(X, Y)))
        append!(newton, @elapsed Newton(100.5))
        append!(lagrange, @elapsed Lagrange(100.5))
        append!(polynomialsPrep, @elapsed (f = fit(X, Y)))
        append!(polynomials, @elapsed f(100.5))
    end
end
```

```julia
using DataFrames

n = repeat(1000:1000:10000,10)
sort!(n)

df = DataFrame(n = n, Newton = newtonPrep, Lagrange = lagrange, Polynomials = polynomialsPrep);
```

```julia
c = combine(
    groupby(df, :n), 
    "Newton" => mean, 
    "Newton" => std,
    "Lagrange" => mean,
    "Lagrange" => std,
    "Polynomials" => mean,
    "Polynomials" => std
)
```

```julia
dfa = DataFrame(n = n, Newton = newton, Polynomials = polynomials);
ca = combine(
    groupby(dfa, :n), 
    "Newton" => mean, 
    "Newton" => std,
    "Polynomials" => mean,
    "Polynomials" => std
)
```

```julia
scatter(legend=:topleft, ylabel="time [s]", xlabel="n", title="Build time")
scatter!(c.n, c.Newton_mean, yerr=c.Newton_std, label="Newton")
scatter!(c.n, c.Lagrange_mean, yerr=c.Lagrange_std, label="Lagrange")
scatter!(c.n, c.Polynomials_mean, yerr=c.Polynomials_std, label="Polynomials")
```

```julia
scatter(ylabel="time [s]", xlabel="n", ylim=(0, 5e-6), title="Access time")
scatter!(ca.n, ca.Newton_mean, yerr=ca.Newton_std, label="Newton")
scatter!(ca.n, ca.Polynomials_mean, yerr=ca.Polynomials_std, label="Polynomials")
```

## Zadanie 5

```julia
function PNeville(X, Y, x)
    n = size(X, 1)
    matrix = Array{Float64}(undef, n, n)
    for i in 1:n
        matrix[i, 1] = Y[i]
    end
    for i in 2:n
        for j in 2:i
            matrix[i, j] = ((x - X[i]) * matrix[i-1, j-1] - (x - X[i - j + 1]) * matrix[i,j-1]) / (X[i - j + 1] - X[i])
        end
    end
    return matrix[n,n]
end;
```

```julia
rng = MersenneTwister(42);
```

```julia
X = 1:10
Y = rand(rng, 10);
```

```julia
PNeville(X, Y, 1)
Neville(x) = PNeville(X, Y, x);
NewtonPrep = coefficients(X, Y)
```

```julia
scatter(X, Y, legend=:topleft, label="data points")
plot!(Neville, extrema(X)..., label="Neville")
plot!(Newton, extrema(X)..., label="Newton")
```

```julia
for x in (1, 3.5, 7.75)
    println("Lagrange:\t", Lagrange(x))
    println("Newton:\t\t", Newton(x))
    println("Neville:\t", Neville(x))
end
```

Wyniki są zbliżone – różnią się tylko na tych miejscach po przecinku, gdzie obliczenia tracą na dokładności

## Zadanie 6

```julia
using Interpolations
```

```julia
rng = MersenneTwister(42);
X = 1:10
Y = rand(rng, 10);
```

```julia
Xf = 1:0.01:10
linear = LinearInterpolation(X, Y)
cubic = CubicSplineInterpolation(X, Y)
quadratic = interpolate(Y, BSpline(Quadratic(Free(OnGrid()))))
Yl = [linear(x) for x in Xf]
Yq = [quadratic(x) for x in Xf]
Yc = [cubic(x) for x in Xf]
f = fit(X, Y);
```

```julia
scatter(X, Y, ylim=(-1, 2), legend=:topleft, label="data points")
plot!(f, extrema(X)..., label="Polynomials")
plot!(Xf, Yl, label="Linear")
plot!(Xf, Yq, label="Quadratic – Free")
plot!(Xf, Yc, label="Cubic")
```
