## Zadanie 1

````julia
using DataFrames
using Statistics
using Plots
````

Mnożenie macierzy – wersja naiwna

````julia
function naive_multiplication(A, B)
    C = zeros(Float64,size(A, 1),size(B, 2))
    for i=1:size(A, 1)
        for j=1:size(B, 2)
            for k=1:size(A, 2)
                C[i, j]=C[i, j]+A[i, k]*B[k, j]
            end
        end
    end
    return C
end;
````

Poprawiona funkcja korzytająca z powyższego oraz z faktu, że można zmieniać kolejność operacji dodawania (a co za tym idzie – kolejność pętli).

````julia
function better_multiplication(A, B)
    C = zeros(Float64, size(A, 1), size(B, 2))
    for j = 1:size(B, 2)
        for k = 1:size(A, 2)
            for i = 1:size(A, 1)
                C[i, j]=C[i, j]+A[i, k]*B[k, j]
            end
        end
    end
    return C
end;
````

````julia
n = repeat(100:100:1000, inner=10)

naive = Array{Float64}(undef, 0)
better = Array{Float64}(undef, 0)
builtin = Array{Float64}(undef, 0)

for i in n
    A = rand(i, i)
    B = rand(i, i)
    append!(naive, @elapsed naive_multiplication(A, B))
    append!(better, @elapsed better_multiplication(A, B))
    append!(builtin, @elapsed A * B)
end

df = DataFrame(n = n, naive = naive, better = better, builtin = builtin)
````

````julia
c = combine(
    groupby(df, :n),
    "naive" => mean,
    "naive" => std,
    "better" => mean,
    "better" => std,
    "builtin" => mean,
    "builtin" => std
)
````

````julia
scatter(legend=:topleft, xlabel="n", ylabel="time [s]", title="Matrix multiplication in Julia")
scatter!(c.n, c.naive_mean, yerr=c.naive_std, label="naive")
scatter!(c.n, c.better_mean, yerr=c.better_std, label="better")
scatter!(c.n, c.builtin_mean, yerr=c.builtin_std, label="built in")
````

## Zadanie 2

Poniżej zamieszczam implementację analogicznego programu w języku C. Będę porównywać ze sobą mnożenie naiwne,
mnożenie lepsze pod względem dostępu do poszczególnyh pól tablicy oraz mnożenie z biblioteki GSL.

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>
#include <gsl/gsl_blas.h>

// A – Matrix n x m
// B – Matrix m x p
// C – Matrix n x p
double **naive_multiplication(double **A, double **B, int n, int p, int m) {
    double **C = (double **) calloc(n, sizeof(double));
    for (int i = 0; i < n; i++)
        C[i] = (double *) calloc(p, sizeof(double));
    
    for (int i = 0; i < n; i++)
        for (int j = 0; j < p; j++)
            for (int k = 0; k < m; k++)
                C[i][j] = A[i][k] * B[k][j];

    return C;
}

// A – Matrix n x m
// B – Matrix m x p
// C – Matrix n x p
double **better_multiplication(double **A, double **B, int n, int p, int m) {
    double **C = (double **) calloc(n, sizeof(double));
    for (int i = 0; i < n; i++)
        C[i] = (double *) calloc(p, sizeof(double));
    
    for (int i = 0; i < n; i++)
        for (int k = 0; k < m; k++)
            for (int j = 0; j < p; j++)
                C[i][j] = C[i][j] + A[i][k] * B[k][j];

    return C;
}

// A – Matrix n x m
void destruct(double **A, int n, int m) {
    for (int i = 0; i < n; i++)
        free(A[i]);
    free(A);
}

double **random_square_matrix(int n) {
    srand(time(NULL));
    double **A = (double **) calloc(n, sizeof(double));
    for (int i = 0; i < n; i++)
        A[i] = (double *) calloc(n, sizeof(double));
    
    for (int i = 0; i < n; i++)
        for (int j = 0; j < n; j++)
            A[i][j] = (double) rand() / RAND_MAX;
    
    return A;
}

double *random_square_array(int n) {
    srand(time(NULL));
    double *A = (double *) calloc(n * n, sizeof(double));
    for (int i = 0; i < n * n; i++)
        A[i] = (double) rand() / RAND_MAX;
    return A;
}

double *zeros_square_array(int n) {
    srand(time(NULL));
    double *A = (double *) calloc(n * n, sizeof(double));
    return A;
}

void print_matrix(double **A, int n, int m) {
    for (int i = 0; i < n; i++) {
        for (int j = 0; j < n; j++)
            printf("%lf ", A[i][j]);
        printf("\n");
    }
    printf("\n");
}

int main(int argc, char **argv) {

    FILE *csv = fopen(argv[1], "w");

    clock_t start_time;
    clock_t end_time;

    for (int n = 100; n <= 1000; n += 100) {
        for (int i = 0; i < 10; i++) {

            // Prepare matrices
            double **A = random_square_matrix(n);
            double **B = random_square_matrix(n);
            double *tC = random_square_array(n);
            double *tD = random_square_array(n);
            double *tE = zeros_square_array(n);
            gsl_matrix_view C = gsl_matrix_view_array(tC, n, n);
            gsl_matrix_view D = gsl_matrix_view_array(tD, n, n);
            gsl_matrix_view E = gsl_matrix_view_array(tE, n, n);

            // Naive
            start_time = clock();
            double **F = naive_multiplication(A, B, n, n, n);
            end_time = clock();
            fprintf(csv, "%d;%lf;", n, (double) (end_time - start_time) / CLOCKS_PER_SEC);
            destruct(F, n, n);

            // Better
            start_time = clock();
            double **G = better_multiplication(A, B, n, n, n);
            end_time = clock();
            fprintf(csv, "%lf;", (double) (end_time - start_time) / CLOCKS_PER_SEC);
            destruct(G, n, n);

            // GSL
            start_time = clock();
            gsl_blas_dgemm(CblasNoTrans, CblasNoTrans, 1.0, &C.matrix, &D.matrix, 0.0, &E.matrix);
            end_time = clock();
            fprintf(csv, "%lf\n", (double) (end_time - start_time) / CLOCKS_PER_SEC);

            // Free memory
            destruct(A, n, n);
            destruct(B, n, n);
            free(tC);
            free(tD);
            free(tE);
        }
    }

    fclose(csv);

    return 0;
}
```

Zamieszczam również plik Makefile, który pomógł wykonane pomiary przekazać do plików csv.

```shell
CC = gcc
CFLAGS = -Wall -Wpedantic

FILE = ./main.c
COMMON = ./common
OPTIMIZED = ./optimized

DATA_COMMON = data-common.csv
DATA_OPTIMIZED = data-optimized.csv

all: run-common run-optimized

run-common: compile-common
	$(COMMON).o $(DATA_COMMON)

run-optimized: compile-optimized
	$(OPTIMIZED).o $(DATA_OPTIMIZED)

compile-common:
	$(CC) $(CFLAGS) $(FILE) -o $(COMMON).o -L/usr/local/lib -lgsl -lgslcblas

```

Można zauważyć, że optymalizacja znacznie skraca zasy wykonania operacji w przypadku funkcji zaimlpementowanych

## Zadanie 4

````julia
using Plots
using Polynomials
````

````julia
func(x) = 1 / (1 + 50x^2);
````

````julia
xs = -1:0.2:1
ys = [func(x) for x in xs];
````

````julia
plot(xlim=(-1.5, 1.5), ylim=(-1, 2), size=(600, 600), title="Runge effect")
scatter!(xs, ys, label="Interpolation points")
plot!(func, label="f(x)")
plot!(fit(xs, ys), extrema(xs)..., label="Polynomial interpolation")
````

````julia
A = zeros(12)
A[12] = 1
xsc = sort(Polynomials.roots(ChebyshevT(A)))
ysc = [func(x) for x in xsc];
````

````julia
plot(xlim=(-1.5, 1.5), ylim=(-1, 2), size=(600, 600), title="Runge effect elimination using Chebyshev polynomial roots")
scatter!(xsc, ysc, label="Interpolation points (Chebyshev roots)")
plot!(func, label="f(x)")
plot!(fit(xsc, ysc), extrema(xs)..., label="Polynomial interpolation")
````

Jak widać, efekt Rungego został wyeliminowany poprzez wyznaczenie punktów interpolacji jako pierwiastków wielomianu Czebyszewa.



## Zadanie 5

````julia
using TaylorSeries
````

````julia
t = Taylor1(Float64, 5)
fun(x) = 1 / (x + 1) / (x + 2)
p = fun(t)
````

````julia
using Polynomials
````

````julia
p = Polynomial(p.coeffs)
pade(x) = Polynomials.PolyCompat.PadeApproximation.Pade(p, 2, 2)(x)
````

````julia
using Plots
````

````julia
limits=(0,5)
plot(xlim=limits, ylim=(-5, 2), title="Taylor and Pade approx comparison")
plot!(fun, limits..., label="1 / ((x-1)(x-2))")
plot!(p, limits..., label="Taylor approximation")
````

````julia
plot!(pade, limits..., label="Pade approximation")
````

Jak widać, aproksymacja funkcji zawierającej osobliwości jest dokładniejsza, gdy użyjemy aproksymacji Pade.
